```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
# Note: Replace 'your_file_path.csv' with the path to your CSV file
file_path = '/content/full_csv_with_haiku'
df = pd.read_csv(file_path)

# Preprocessing steps (as needed based on the initial analysis provided)
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Sorting the DataFrame by date for correct plotting
clean_df_sorted = df.sort_values(by='date')
```

## Matriz de correlacion!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'clean_df_sorted' is your DataFrame that has been sorted and cleaned

# Selecting the columns relevant for correlation analysis
correlation_columns = [
    'Price_COL',
    'Price_USA',
    'title_sentiment_normalized',
    'subtitle_sentiment_normalized',
    'content_sentiment_normalized',
    'Title_haiku_simple',
    'Subtitle_haiku_simple',
    'Content_haiku_simple',
    'Title_haiku_sbs',
    'Subtitle_haiku_sbs',
    'Content_haiku_sbs'
]

# Calculating the correlation matrix for the selected columns
correlation_matrix = clean_df_sorted[correlation_columns].corr()

# Visualizing the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5, fmt=".2f")
plt.title('Correlation Matrix between Price Changes and Analysis Scores')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'clean_df_sorted' is your DataFrame

# Calculating the correlation of Price USA & COL with each analysis score separately
analysis_scores = [
    'title_sentiment_normalized',
    'subtitle_sentiment_normalized',
    'content_sentiment_normalized',
    'Title_haiku_simple',
    'Subtitle_haiku_simple',
    'Content_haiku_simple',
    'Title_haiku_sbs',
    'Subtitle_haiku_sbs',
    'Content_haiku_sbs'
]

# Create a new DataFrame for correlation values
correlations = pd.DataFrame(index=['Price_COL', 'Price_USA'], columns=analysis_scores)

# Calculating correlations for each score with Price_COL and Price_USA
for score in analysis_scores:
    for price in ['Price_COL', 'Price_USA']:
        correlation = clean_df_sorted[[price, score]].corr().iloc[0,1]
        correlations.loc[price, score] = correlation

# Converting correlation values to numeric for plotting
correlations = correlations.astype(float)

# Visualizing the focused correlation matrix
plt.figure(figsize=(14, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', linewidths=.5, fmt=".2f")
plt.title('Correlation between Price (USA & COL) and Analysis Scores')
plt.yticks(rotation=0)  # Keep the y labels readable
plt.show()
```

It seems like there is a high level of correlation (relatively) between the PRICE_USA and the PRICE_COL
